```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd




# DATA PREPROCESSING
df = pd.read_csv(r"give your path")
df.drop_duplicates(inplace=True)
df.isnull().sum()/len(df)
df.dropna(inplace=True)
df = pd.get_dummies(df,columns=['category'],dtype='int')
df = pd.get_dummies(df,columns=['country'],dtype='int')
df = pd.get_dummies(df,columns=['device'],dtype='int') # one hot encoding

# FEATURE ENGINEERING
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df['second'] = df['date'].dt.second
df.drop(columns=['minute','second','hour'],inplace=True)
df.drop(columns=['date'], inplace=True)
df['video_id'] = df['video_id'].str.replace('vid_', '', regex=False).astype(int)

# SCALING DATA
scaler = StandardScaler()
df[['watch_time_minutes']] = scaler.fit_transform(df[['watch_time_minutes']])
scaler = MinMaxScaler() # scale your data in range of 0-1
# x-min/(max-min)
df[['views','likes','video_length_minutes','subscribers','comments']] = scaler.fit_transform(df[['views','likes','video_length_minutes','subscribers','comments']])


# TRAINING MODEL
X = df.drop(['ad_revenue_usd'], axis=1)
y = df['ad_revenue_usd'].astype(int)

feature_cols = X.columns.tolist()

# save for use in app
import joblib
joblib.dump(feature_cols, "feature_cols.pkl")

# train & save model
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X, y)
joblib.dump(model, "yt_ad_revenue_model.pkl")







sample = df.drop(['ad_revenue_usd'], axis=1).iloc[[1,2,3,4]]
pred = model.predict(sample)
print(pred[3])
```
